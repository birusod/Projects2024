## Setup

### Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

### Data

```python
url_main = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/"
url_data = "data/2024/2024-08-06/olympics.csv"
rdt = pd.read_csv(url_main + url_data)
```

## Data Wrangling

```python
rdt.head(3)
```

### Summer games

```python
dfs = rdt[rdt['season'] == 'Summer'].dropna()
dfs.describe()
```

```python
dfs.height.hist()
```

```python
dfs.weight.hist()
```

```python
sns.histplot(x = 'height', data=dfs)
plt.show()
```

### Height vs Weight by season

```python
sns.pairplot(data=dfs[['height', 'weight']])
plt.show()
```

```python
sns.scatterplot(x='height', y = 'weight', hue = 'season',  data=rdt)
plt.show()
```

### Age vs Weight season

```python
sns.scatterplot(x  ='weight', y = 'age', data=rdt)
plt.show()
```
