### Member appearance on issues between 1960 and 1990

```python
ptitle =  "X-MEN MUTANT MONEYBALL\n"
psubtitle = "Total number of issues each X-Men member \nappeared between 1960s and 1990\n"
```

```python
appearance_df  = xm.groupby(['name'])['appearance'].sum().sort_values(ascending=False)
appearance_df.plot.bar(
    title = ptitle + psubtitle)
plt.show()
```

```python
appearance_sorted  = (
    xm.groupby(['name'])['appearance'].sum()
    .reset_index()
    .sort_values('appearance', ascending = False)
)

fig, ax = plt.subplots(figsize= (6, 8))

ax  = sns.barplot(
    appearance_sorted,
    x = 'appearance', y = 'name',
    color = '#6F00FF',
    height=.8 #bar width
    )
ax.set(xlabel="", ylabel="")
ax.tick_params(left=False)

fig.suptitle(ptitle, x = 0.5, y = .95, fontsize = 16, color = 'purple', fontweight = 'bold')
fig.text(x=0.5, y=.87, 
         s= psubtitle, fontsize=10, alpha=0.5,
         horizontalalignment='center',
        #verticalalignment='center',
        )

#g.xaxis.set_major_formatter('{x:.0f}Mlns')
#g.xaxis.set_major_formatter('${x:.0f}')
plt.show()

```
